```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Import csv into DataFrame
housing_data_df = pd.read_csv('Resources/austinHousingData.csv')
```

```python
# Make a working copy of DataFrame
df = housing_data_df.copy()
df.head()
```

```python
# Display all columns of the DataFrame 
pd.options.display.max_columns = None
display(df)
```

```python
# Check data types
df.info()
```

```python
# keep relevant columns
clean_df = df[['zpid','city','zipcode',
               'homeType','yearBuilt',
               'latestPrice','numPriceChanges', 
               'avgSchoolDistance']]

clean_df['Outcome'] = np.where(clean_df['zipcode']==78748, 1, 0)
clean_df
```

```python
zip_code = [78748, 78745, 78749, 78704]


for code in zip_code:
    if code in clean_df['zipcode']:
        clean_df['Outcome'] = np.where(clean_df['zipcode']==code, 1, 0)
clean_df
```

```python
clean_df.loc[clean_df['Outcome']==1]
```

```python
clean_df.zipcode.value_counts()
```

```python
# Drop any null values
clean_df.dropna()
```

```python
# Drop any duplicate values
clean_df.drop_duplicates()
```

```python
# Suppress Scientific Notation
pd.options.display.float_format = '{:.2f}'.format

# Group average home price by zipcode
avg_price = clean_df.groupby(['zipcode']).mean()['latestPrice']
avg_price
```

```python
# Create new DataFrame with zipcodes and average home price
avg_df = pd.DataFrame({
    'Average Price' : avg_price
})

# Sort the new DataFrame
avg_df.sort_values(by='Average Price', ascending=True)
```

```python
# View average price change by zipcode
avg_price_change = clean_df.groupby(['zipcode']).mean()['numPriceChanges']
avg_price_change
```

```python
# Create DataFrame with just Austin homes
austin_df = clean_df.loc[clean_df['city']=='austin']
austin_df
```

```python
austin_df.zipcode.value_counts()
```

```python
# Create DataFrame with surrounding areas
suburb_df = clean_df.loc[clean_df['city']!='austin']
suburb_df
```
